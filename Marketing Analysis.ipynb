```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("marketing_data.csv")
df.head()
```

## Exploratory Data Analysis

```python
df.info()
```

```python
df[' Income ']=df[' Income '].str.replace("$","")
df[' Income ']=df[' Income '].str.replace(",","").astype('float')
```

```python
df.head()
```

### Checking Null Values and Outliers

```python
df.isnull().sum()
```

```python
df[" Income "]=df[" Income "].fillna(df[" Income "].median())
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
import numpy as np
df_check=df.copy()
df_check=df.drop(columns=['ID','AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1',
       'AcceptedCmp2', 'Response', 'Complain']).select_dtypes(include=np.number)
```

```python
df_check.head()
```

```python
import seaborn as sns
fig, ax = plt.subplots(4, 4, figsize = (15, 16))
ax=ax.flatten()

for i, c in enumerate(df_check):
    sns.boxplot(x = df_check[c], ax = ax[i])
    
plt.suptitle('Outlier Analysis using BoxPlots', fontsize = 25)
fig.tight_layout()
```

### From the Boxplots we see Year_Birth and Income have a Few Outliers. We will change Year_Birth but ignore Income because there exists high earners.

```python
plt.figure(figsize=(8,4))
sns.histplot(df['Year_Birth'])
plt.title('Year Birth Distribution', size=16)
plt.ylabel('Count');
```

```python
#Change Anybody born to the Median Year of Birth
df['Year_Birth']=df['Year_Birth'].apply(lambda x:df['Year_Birth'].median() if x<=1900 else x)
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['Year_Birth'])
plt.title('Year Birth Distribution', size=16)
plt.ylabel('Count');
```

```python
#Change from float to integer
df['Year_Birth']=df['Year_Birth'].astype(int)
```

### Changing Variables which are needed Transformation?

```python
import datetime
year=datetime.datetime.today().year
df['Age']=year-df['Year_Birth']
```

```python
df.head()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['Age'],kde=True)
plt.title('Age Distribution',size=16)
```

### Instead of having customers categorized by the full date they signed up with the company, it can be simplified to just year

```python
df['Year_customer']=pd.DatetimeIndex(df['Dt_Customer']).year
```

```python
df.head()
```

```python
df.columns
```

```python
## Kidhome and Teenhome can be combined to create Dependents
df['Dependents']=df['Kidhome']+df['Teenhome']

#Total Amount Spent
mnt_cols=[col for col in df.columns if 'Mnt' in col]
df['TotalMnt']=df[mnt_cols].sum(axis=1)

#Total Purchases
purch_cols=[col for col in df.columns if 'Purchases' in col]
df['Total_Purchases']=df[purch_cols].sum(axis=1)

#Total Campaigns Accepted
campaigns_cols=[col for col in df.columns if 'Cmp' in col]+['Response']
df['Total_Campaigns']=df[campaigns_cols].sum(axis=1)
```

```python
df['Year_Birth']=df['Year_Birth'].astype(int)
```

```python
# #we will drop the columns that we transformed
# # df=df.drop(df['Year_Birth'],axis=1)

# df=df.drop(["Kidhome", "Teenhome"],axis=1)

# df=df.drop(df['Dt_Customer'],axis=1)
```

```python
df_group=df.copy()
age_bins=[20,30,40,50,60,120]
age_labels=["20s","30s","40s","50s","60+"]
df_group["Age_group"]=pd.cut(df_group.Age,age_bins,labels=age_labels,include_lowest=True)

income_bins=[0,19000,39999,59999,89999,200000]
income_labels=["Very Low","Low","Middle","High","Very High"]
df_group["Income_group"]=pd.cut(df_group[" Income "],income_bins,labels=income_labels,include_lowest=True)
```

```python
df_group.drop([" Income ","Age"],axis=1)
```

```python
f,ax=plt.subplots(2,2,figsize=(14,10))
sns.barplot(x='Income_group',y='TotalMnt',data=df_group,ax=ax[0][0])
ax[0][0].set_title('Income Group to Total Spent',fontweight="bold")

sns.barplot(x='Dependents',y='TotalMnt',data=df_group,ax=ax[0][1])
ax[0][1].set_title('Dependents to Total Spent',fontweight="bold")

sns.barplot(x='Education',y="TotalMnt",data=df_group,ax=ax[1][0])
ax[1][0].set_title('Education Level to Total Spent',fontweight="bold")

sns.barplot(x='Education',y=" Income ",data=df_group,ax=ax[1][1])
ax[1][0].set_title('Income to Education',fontweight="bold")

# sns.barplot(x='')
```

# Statstical Analysis

```python
corr=df.drop(columns="ID").select_dtypes(include=np.number).corr(method='kendall')
fig,ax=plt.subplots(1,1,figsize=(16,16),dpi=200)
sns.heatmap(corr,ax=ax,annot=True,fmt='.2f',square=True)
```

Based on the correlation matrix ,the factors that are significantly related to thr number of store purchases are:
    1. Income
    2.MntWines
    3.MntMeatPRoducts
    4.NumCatalogPurchases
    5.TotalMnt
    6.Total_Purchases

### Does US fare significantly better than the Rest of the World in terms of total purchases?

```python
plt.figure(figsize=(4,4))
df.groupby('Country')['Total_Purchases'].sum().sort_values(ascending=False).plot(kind='bar')

plt.title("Total Number Of Purchases By Country",size=16)
```

### Fish has Omega 3 fatty acids which are good for the brain. Accordingly, do "Married PhD candidates" have a significant relation with amount spent on fish? What other factors are significantly related to amount spent on fish? 

```python
df2=df.copy()
# df2['Married_Phd']=df2.loc[(df2['Marital_Status']=='Married') & (df2['Education']=='PhD')]
df2['Married_Phd']=df2['Marital_Status']+df2['Education']
df2['Married_Phd']=df2['Married_Phd'].apply(lambda x:1 if x=='MarriedPhD' else 0)

plt.figure(figsize=(4,4))
sns.boxplot(x='Married_Phd',y='MntFishProducts',data=df2)
```

```python
df2 = df2.drop(columns = ['ID', 'Education', 'Marital_Status', 'AcceptedCmp1', 'AcceptedCmp2', 
                          'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response', 
                          'Complain', 'Year_customer'], axis = 1)
```

```python
corr2=df2.corr(method='pearson')
fig,ax=plt.subplots(1,1,figsize=(16,16),dpi=200)
sns.heatmap(corr2,ax=ax,annot=True,fmt='.2f',square=True)
plt.show()
```

### From the heatmap, we can see that the factors that influence the purchase of fish are:

1-MntWines
2-MntMeatProducts
3-MntSweetProducts
4-TotalMnt
5-NumCatalPurchases

## Which Product is Performing Best?

```python
product=pd.DataFrame(round(df[['MntWines','MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
                               'MntGoldProds']].mean(),1),
                    columns=['Average']).sort_values(by='Average').reset_index()
plt.figure(figsize=(8,4))
plt.xticks(rotation=45)
sns.barplot(x='index',y='Average',data=product)
```
